## Import data and preliminary data manipulation

First, we removed the duration feature since it is prune to data leakage as referred by the authors.

Then, we extend the feature **month** by adding its numeric value besides its categorical representation.

```python
%matplotlib inline

import pandas as pd
import numpy as np

# Read the data frame
data_filename = 'bank.csv'
df = pd.read_csv(data_filename, sep=';')

# Remove the duration column since it is prune to data leakage
df = df.drop('duration', 1)

# Transform the target variable to numeric and remove it from the dataframe
y = (df.y == 'yes').astype(int)
df = df.drop('y', 1)

# Transform the month to numeric value
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
months = {m: i + 1 for i, m in enumerate(months)}
df['num-month'] = np.asarray([months[m] for m in df.month])

df.head()
```

## Analysis of numeric variables

```python
from sklearn.feature_selection import mutual_info_classif
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import seaborn as sns

numeric_variables = ['age', 'balance', 'num-month', 'day', 'pdays', 'previous', 'campaign']

correlation = [spearmanr(df[nv], y)[0] for nv in numeric_variables]

print('Variable Spearman-rank')
for var, corr in sorted(zip(numeric_variables, correlation), key=lambda x: abs(x[1])):
    print('%15s %7.4f' % (var, corr))
```

**Spearman**

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (5.0, 2.0)

for nv in numeric_variables:
    plt.title(nv)
    plt.hist(df[nv].values[y == 1], 20, color='blue', alpha=0.5, normed=True)
    plt.hist(df[nv].values[y == 0], 20, color='red', alpha=0.5, normed=True)
    plt.show()
```

The features **balance**, **pdays**, **previous** and **campaign** have outliers. Let's inspect.

```python
df[numeric_variables].describe()
```

Apply log transform to reduce outlier impact.

Since the **pdays** variable encodes missing values as **-1**, we created an additional binary variable with a flag of missing vs. non-missing and leave the missing values as actual missing values.

```python
new_balance = df.balance.values
new_balance[new_balance < 0] = -np.log(1 - new_balance[new_balance < 0])
new_balance[new_balance > 0] = np.log(1 + new_balance[new_balance > 0])

new_pdays = (df.pdays.values).astype(float)
new_pdays[df.pdays != -1] = np.log(1 + df.pdays[df.pdays != -1].values)
new_pdays[df.pdays == -1] = np.nan

df['balance'] = new_balance
df['pdays'] = new_pdays
df['pdays-missing'] = ~np.isnan(new_pdays)

df.previous = np.log(1 + df.previous.values)
df.campaign = np.log(1 + df.campaign.values)

df[numeric_variables].describe()

for nv in numeric_variables:
    plt.title(nv)
    values = df[nv].values
    plt.hist(values[(y == 1) & (~np.isnan(values))], 20, color='blue', alpha=0.5, normed=True)
    plt.hist(values[(y == 0) & (~np.isnan(values))], 20, color='red', alpha=0.5, normed=True)
    plt.show()
```

## Analysis of categorical variables

```python
categorical_variables = ['job', 'marital', 'education', 'default',
                         'housing', 'loan', 'contact',
                         'poutcome', 'month']

for cvi, cv in enumerate(categorical_variables):
    plt.title(cv)
    values = df[cv].values
    uvalues = list(np.unique(values))
    values = np.asarray([uvalues.index(x) for x in df[cv]])

    nvalues = len(uvalues)
    h = plt.bar(range(nvalues), [np.sum(values[y == 1] == v) / np.sum(y == 1) for v in range(nvalues)],
                label=uvalues, color='blue', alpha=0.4)
    h = plt.bar(range(nvalues), [np.sum(values[y == 0] == v) / np.sum(y == 0) for v in range(nvalues)],
                label=uvalues, color='red', alpha=0.4)

    xticks_pos = [0.65*patch.get_width() + patch.get_xy()[0] for patch in h]
    plt.xticks(xticks_pos, uvalues,  ha='right', rotation=45)
    plt.show()
```
